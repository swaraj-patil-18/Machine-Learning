## Unemployment Rate Analysis: Data Visualization

Estimated number of employees according to different regions of India

```python
df.columns= ["States","Date","Frequency",
               "Estimated Unemployment Rate","Estimated Employed",
               "Estimated Labour Participation Rate","Region",
               "longitude","latitude"]
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Employed",hue="Region",data=df)
plt.show()               
```

Unemployment rate according to different regions of India

```python
plt.figure(figsize=(12, 10))
plt.title("Indian Unemployment")
sns.histplot(x="Estimated Unemployment Rate",hue="Region",data=df)
plt.show()   
```

```python
df['int_column'] = df['Estimated Unemployment Rate'].astype(int)
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year


df['month'] = df['Date'].dt.month_name().str.slice(stop=3)


```

```python
df['year'].value_counts()
```

Lineplot:Unemployment Rate by Region and Month

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a line plot with one line per region
sns.lineplot(data=df, x='month', y='int_column', hue='Region')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate by Region and Month')
```

Barplot:Unemployment Rate by Region and Month

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a bar plot with one bar per year
sns.barplot(data=df, x='month', y='int_column', hue='Region')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate by Region and Month')
```

**Area plot with one area per region to show the proportion of each region's unemployment rate over time.**

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create an area plot with one area per region
sns.lineplot(data=df, x='month', y='int_column', hue='Region', estimator=None, alpha=0.2)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Unemployment Rate')
plt.title('Unemployment Rate by Region and Month')
```
